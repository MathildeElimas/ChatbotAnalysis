### Boosting model

#### Train/Test

```python
Y = [output_tokenizer.word_index[intent.lower()] for intent in intent]
```

```python
if method == 'tfidf':
    Tfidf_vect = TfidfVectorizer(max_features=5000, vocabulary = list(vocab.keys()))
    Tfidf_vect.fit(cleaned_words)

    X = Tfidf_vect.transform(cleaned_words)
elif method == 'count':
    Count_vect = CountVectorizer(max_features=5000, vocabulary = list(vocab.keys()))
    Count_vect.fit(cleaned_words)

    X = Count_vect.transform(cleaned_words)
```

```python
Train_X, Test_X, Train_Y, Test_Y = train_test_split(X, Y,test_size=0.2)
```

#### Model definition & training

```python
k_fold = KFold(n_splits=100)
```

```python
# Classifier - Algorithm - SVM
# fit the training dataset on the classifier
boost = RandomForestClassifier(max_depth = 500)

cv = cross_validate(boost, X, Y, cv = k_fold, scoring = 'balanced_accuracy')
boost.fit(Train_X,Train_Y)
```

```python
with open(os.getcwd() + '\\resultats\\boosting_' + method +'.pkl','wb') as f:
    pickle.dump(boost,f)
```

```python
cv['test_score'].mean()
```

#### Model evaluation

Test with test dataset

```python
# predict the labels on validation dataset
predictions_boosting = boost.predict(Test_X)
# Use accuracy_score function to get the accuracy
print("Boosting Accuracy Score -> ", balanced_accuracy_score(Test_Y, predictions_boosting)*100)
```
