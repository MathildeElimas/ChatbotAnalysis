```python
import pandas as pd
```

```python
PATH = 'C:\\Users\\MathildeElimas\\Documents\\BYTEL\\Bot_Data\\'
```

```python
riche_comment = pd.read_csv(PATH + 'bulkQuery_result_riche.csv', sep = ',')
```

```python
riche_comment.head()
```

```python
riche_comment['CommentRichBody__c'] = riche_comment['CommentRichBody__c'].str.replace('&quot;','\"')
riche_comment['CommentRichBody__c'] = riche_comment['CommentRichBody__c'].str.replace('null', 'None')
riche_comment['CommentRichBody__c'] = riche_comment['CommentRichBody__c'].str.replace('true', 'True')
riche_comment['CommentRichBody__c'] = riche_comment['CommentRichBody__c'].str.replace('false', 'False')
```

```python
riche_comment['CommentRichBody__c'] = riche_comment['CommentRichBody__c'].apply(lambda x : eval(x) if x == x else None)
```

```python
test = pd.json_normalize(riche_comment['CommentRichBody__c'])
```

```python
riche_comment = pd.merge(riche_comment, test, left_index=True, right_index=True)
```

```python
riche_comment.head()
```

```python
del(riche_comment['isBot'])
del(riche_comment['delay'])
del(riche_comment['CommentRichBody__c'])
```

```python
riche_comment.rename(columns={'ParentId__c' :'ParentId'},
               inplace=True)
```

```python
riche_comment.to_csv(PATH + 'bulkQuery_result_riche_propre.csv')
```
